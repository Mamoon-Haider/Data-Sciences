# Now data is ready for Feature Scaling 


# Lets seperate the Train data from the combined dataset

```python
df['Purchase'].isnull()
```

```python
df_test = df[df['Purchase'].isnull()]
df_test
```

```python
df_train = df[~df['Purchase'].isnull()]
df_train
```

# We will use simple Train-Test split cross validation technique

# For "X", we will remove "Dependent variable: Purchase" from the Train_dataset.

# For "Y", we will just extract the "Dependent variable: Purchase" from the Train_dataset.

```python
X = df_train.drop("Purchase", axis = 1)
X
```

```python
Y = df_train['Purchase']
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split (X, Y, test_size=0.30, random_state=42)
```

# Now, next step is to perform Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_test
```

```python
X_train
```

# Cross Validation on X_train dataset

```python
# Importing model
from sklearn.linear_model import LinearRegression
# Importing cross validation
from sklearn.model_selection import cross_val_score
```

# lets define model first

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

# Lets give parameters to cross-val score

```python
mse = cross_val_score(regression,X_train, y_train, scoring= 'neg_mean_squared_error', cv = 10 )
mse
```

# Lets predict the test data

```python
regression_predict = regression.predict(X_test)

regression_predict
```

# lets check the accuracy of preidction

```python
import seaborn as sns
sns.displot(regression_predict - y_test, kind = 'kde' )
```

# Variation between actual and predicted values is very high, going between -10,000 to 10,000. Not very good.

```python
y_predict = regression_predict

Y_actual = y_test
```

```python
from sklearn.metrics import r2_score

score = r2_score(regression_predict, y_test)

score
```

# Applying Ride Regression      &     Grid Search Cross Validation Technique

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

# In ridge, alpha is the parameter that we need to decide. 
# Let's use the GridSearchCV to get the best parameter.

```python
ridge_regressor = Ridge()
```

```python
ridge_regressor
```

# Lets define the parameter

```python
parameters = {'alpha': [1,2,5,20,40,60,80]}
ridgecv = GridSearchCV(ridge_regressor, parameters, scoring = 'neg_mean_squared_error', cv = 5)
ridgecv.fit(X_train, y_train)
```

# let's check the best Hyperparameter

```python
print(ridgecv.best_params_)
```

```python
print(ridgecv.best_score_)
```
